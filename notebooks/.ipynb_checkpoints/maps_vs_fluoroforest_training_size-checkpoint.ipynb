```python
# the core random forest functionality is used below, it reamins the same as with the full workflow, but without the need for annotation and setting up other class structures
```

```python
def calculate_metrics(y_true, y_pred, class_names = None):

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    classes = np.unique(np.concatenate((y_true, y_pred)))
    
    per_class_metrics = {}
    
    # per class
    for c in classes:
        is_class = y_true == c
        is_predicted_class = y_pred == c
        
        tp = np.sum(is_class & is_predicted_class)
        fp = np.sum(~is_class & is_predicted_class)
        fn = np.sum(is_class & ~is_predicted_class)
        
        tn = np.sum(~is_class & ~is_predicted_class)

        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        
        # f1 score
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        
        # accuracy = (TP+TN) / num samples
        accuracy = (tp + tn) / len(y_true) if len(y_true) > 0 else 0
        
        # metrics for current class
        per_class_metrics[c] = {
            'precision': precision,
            'recall': recall,
            'f1-score': f1_score,
            'accuracy': accuracy,
            'support': np.sum(is_class)
        }
    

    metrics_df = pd.DataFrame.from_dict(per_class_metrics, orient='index')

    # overall metrics for precision, recall, and f1
    macro_precision = metrics_df['precision'].mean()
    macro_recall = metrics_df['recall'].mean()
    macro_f1_score = metrics_df['f1-score'].mean()
    
    # accuracy
    overall_correct = np.sum(y_true == y_pred)
    overall_accuracy = overall_correct / len(y_true) if len(y_true) > 0 else 0
    
    metrics_df.loc['overall'] = [macro_precision, macro_recall, macro_f1_score, overall_accuracy, None]

    if class_names is not None:
        name_map = {label: name for label, name in enumerate(class_names)}
        
        new_index = [name_map.get(i, i) for i in metrics_df.index]
        metrics_df.index = new_index

    return metrics_df
```

```python
# class names and color palette 
class_names = ['B', 'CD4 T', 'CD8 T', 'DC', 'Endothelial', 'Epithelial', 
               'Lymphatic', 'M1', 'M2', 'Mast', 'Monocyte', 'NK', 'Neutrophil', 
              'Other', 'Treg', 'Tumor']

# extracting ground truth for model performance
gt_labels = maps_validation['cell_label'].to_numpy()
```

# Random forest run on same data for MAPS dataset sampling test

```python
i = 0

model_results = []
for k,v in loaded_indices.items():
    print(k)
    for k2,v2 in v.items():

        i+=1 # change seed through loops
        current_sample_indices = v2
        # prepare X_train, y_train for the current fold
        X_train_fold = maps_train.iloc[v2, :-1]
        y_train_fold = maps_train.iloc[v2, -1]
    
        # X_test, y_test are fixed (validation set)
        X_test_fixed = maps_validation.iloc[:, :-1]
        y_test_fixed = maps_validation.iloc[:, -1]
    
        # random state for reproducibility, updated hyperparameters -- see hyperparameter tuning script
        rf = RandomForestClassifier(n_estimators = 250, max_features = 'sqrt',
                                    max_depth = 10, min_samples_leaf = 2, 
                                    min_samples_split = 5, random_state = 5 + i)
        
        rf.fit(X_train_fold, y_train_fold)
        y_pred_fold = rf.predict(X_test_fixed)
    
        y_true_eval = y_test_fixed # Renaming for clarity
        
        # metrics
        metrics_report = calculate_metrics(gt_labels, y_pred_fold, class_names = class_names)
        metrics_report['samples_per_class'] = k.split('_')[1]
        metrics_report['fold'] = k.split('_')[1]

        model_results.append(metrics_report) # concatentate final
    
```

```python
# read in from maps final data (trained on same, validated on same)
```

```python
final_results = pd.concat(model_results, axis = 0)
```

### Below models were compared to see at which point MAPS begins to outperform random forests via F1-macro scores. In most casess it took 100+ cells before it peformed better. 

```python
combined_df = combined_df[combined_df['class'] != 'overall']

# average across folds for each model
mean_f1 = (
    combined_df
    .groupby(['class', 'samples_per_class', 'model'], as_index=False)['f1-score']
    .mean()
)

mean_f1['samples_per_class'] = mean_f1['samples_per_class'].astype(int)

maps_df = mean_f1[mean_f1['model'] == 'MAPS'].rename(columns={'f1-score': 'MAPS'})
rf_df = mean_f1[mean_f1['model'] == 'RF'].rename(columns={'f1-score': 'RF'})

merged = pd.merge(
    maps_df[['class', 'samples_per_class', 'MAPS']],
    rf_df[['class', 'samples_per_class', 'RF']],
    on=['class', 'samples_per_class'],
    how='outer'
)

merged = merged.sort_values(['class', 'samples_per_class']).reset_index(drop=True)
```

```python
def find_crossover(df):
    df = df.sort_values('samples_per_class')
    df = df.dropna(subset=['MAPS', 'RF'])  
    crossover = df[df['MAPS'] > df['RF']]
    if not crossover.empty:
        return crossover.iloc[0][['samples_per_class', 'MAPS', 'RF']]
    return pd.Series({'samples_per_class': None, 'MAPS': None, 'RF': None})

crossover_points = merged.groupby('class', group_keys=False).apply(find_crossover).reset_index()
crossover_points
```

```python
df_sorted = crossover_points.sort_values(by='samples_per_class', ascending=False)

plt.figure(figsize=(6, 3))
sns.barplot(x='samples_per_class', y='class', data=df_sorted, color='lightgray', edgecolor = 'black', alpha = 0.5)

plt.xlabel('Samples per class (crossover point)', fontsize=12)
plt.ylabel('Cell Type', fontsize=12)
plt.grid(False)
plt.tight_layout()
plt.show()
```
