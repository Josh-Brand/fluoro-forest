### Comparison of performance via precision-recall cuves at 30 training samples per model

```python
# requires cell_annotation environment
import sys
import os
import io
import json
import importlib
import numpy as np
import collections
import scipy
import sklearn

import seaborn as sns
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
```

```python
# reading in util functions:
# notebook directory
current_dir = os.getcwd()

# project directory
root_dir = os.path.abspath(os.path.join(current_dir, '..', '..'))
os.chdir(root_dir)

# for importing utils
sys.path.append(os.path.join(root_dir, 'src', 'functions'))
```

```python
maps_pr = []
maps_path = os.path.join(root_dir, 'maps', 'outs')
os.listdir(maps_path)

for f in os.listdir(maps_path):
    if 'pr_curve_data' in os.path.join(maps_path, f):
        fle = os.path.join(maps_path, f)
        print(fle)
        pr = pd.read_csv(fle, low_memory = False)
        pr['model'] = 'MAPS'
        maps_pr.append(pr)
```

```python
rf_pr = []
rf_path = os.path.join(root_dir, 'other_data', 'pr_curves')
os.listdir(rf_path)

for f in os.listdir(rf_path):
    if 'pr_curve_data' in os.path.join(rf_path, f):
        fle = os.path.join(rf_path, f)
        print(fle)
        pr = pd.read_csv(fle, low_memory = False)
        pr['model'] = 'RandomForest'
        rf_pr.append(pr)
```

```python
merged_pr = pd.concat([pd.concat(rf_pr, axis = 0), 
                       pd.concat(maps_pr, axis = 0)])
```

```python
merged_pr.head
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from sklearn.metrics import PrecisionRecallDisplay

df_filtered = merged_pr[
    (merged_pr['curve_type'] == 'per_class') & 
    (merged_pr['fold'] == 1)
].copy()

models = sorted(df_filtered['model'].unique())
class_names = sorted(df_filtered['class_name'].unique())
num_classes = len(class_names)
num_models = len(models)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
axes = axes.flatten()

class_palette = cm.get_cmap('tab20', num_classes)

# compare 2 models (each of 5 cv are comparable)
for i, model_name in enumerate(models):
    ax = axes[i]
    
    # iterate classes
    for j, class_name in enumerate(class_names):
        # filter to class
        plot_data = df_filtered[
            (df_filtered['model'] == model_name) &
            (df_filtered['class_name'] == class_name)
        ].copy()
        
        if not plot_data.empty:
            precision_points = plot_data['precision'].values
            recall_points = plot_data['recall'].values
            
            # plot each curve
            display = PrecisionRecallDisplay(
                precision=precision_points,
                recall=recall_points
            )
            
            color = class_palette(j)
            label = f"{class_name}"
            
            display.plot(ax=ax, name=label, color=color, alpha=0.8)
    
    ax.set_title(f'Precision-Recall Curves for {model_name}', fontsize=14)
    ax.set_xlabel('Recall')
    ax.set_ylabel('Precision')
    # ax.legend(loc="lower left", fontsize=0, handlelength=1.5, frameon=False)
    ax.get_legend().remove()
    ax.grid(False)
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import auc

df_per_class = merged_pr[merged_pr['curve_type'] == 'per_class'].copy()

grouped_data = df_per_class.groupby(['model', 'class_name', 'fold'])

ap_values = []

for name, group in grouped_data:
    model_name, class_name, fold = name
    
    if 'recall' in group.columns and 'precision' in group.columns:
     
        group_sorted = group.sort_values(by='recall')
        
        recall_points = group_sorted['recall'].values
        precision_points = group_sorted['precision'].values
        
       
        if len(recall_points) > 1 and len(precision_points) > 1:
            average_precision = auc(recall_points, precision_points)
            ap_values.append({'model': model_name, 'class_name': class_name, 'average_precision': average_precision})

ap_df = pd.DataFrame(ap_values)
comparison_table = ap_df.groupby(['class_name', 'model'])['average_precision'].mean().unstack()

comparison_table = comparison_table.round(4)
```

```python
new_names = ['fluoroforest' if 'RandomForest' in x else 'maps' for x in ap_df['model']]
ap_df['new_names'] = new_names
```

```python
plt.figure(figsize=(6, 4))

sns.barplot(
    data=ap_df,
    x='class_name',
    y='average_precision',
    hue='new_names',
    palette='tab10',
    errorbar=None,     
    width=0.7,
    legend=False,
    hue_order=['fluoroforest', 'maps'],
    edgecolor = 'black',
    alpha=0.6 
)


sns.swarmplot(
    data=ap_df,
    x='class_name',
    y='average_precision',
    hue='new_names',
    dodge=True,        
    palette='tab10',
    legend=False,      
    size=3,
    hue_order=['fluoroforest', 'maps'],
    edgecolor='black',
    linewidth=1,
    marker='o', 
    alpha = 0.7
)


plt.xticks(rotation=45, ha='right')

# Set title and labels
plt.xlabel("Cell Type", size = 13)
plt.ylabel("Average Precision", size = 13)
plt.ylim(bottom=0)


# Adjust the layout to make room for rotated labels
plt.tight_layout()
plt.show()
```
