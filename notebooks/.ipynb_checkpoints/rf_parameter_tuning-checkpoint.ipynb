## Random forest hyperparameter tuning
This was done to get a better idea on which parameters are most likely to improve performance. 
We use the MAPS dataset as an example since there is plenty to work with without risk of data leakage.

```python
import sys
import os
import io
import json
import importlib
import numpy as np
import collections
import scipy
import sklearn
import time

import seaborn as sns
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt


from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.ensemble import BalancedRandomForestClassifier
```

```python
# only original maps training data used as to avoid bias in other training/validation work

current_dir = os.getcwd()

# project directory
root_dir = os.path.abspath(os.path.join(current_dir, '..', '..'))
os.chdir(root_dir)
```

```python
maps_data = pd.read_csv(os.path.join(root_dir, 'maps', 'data', 'cell_phenotyping', 'train.csv')) # taken from maps module
```

```python
# getting an idea of cell type distributions and how we would need to subsample for balanced training
maps_ds = maps_data.iloc[1:1000,:]
maps_ds['cell_label'].value_counts()
```

```python
# sampling for each class
MIN_SAMPLES = 40

sampled_data_list = []

grouped = maps_data.groupby('cell_label')

for label, group in grouped:
    class_size = len(group)

    if class_size >= MIN_SAMPLES:
        sampled_group = group.sample(n = MIN_SAMPLES, random_state=42)
    else:
        print(f"Warning: Class {label} only has {class_size} samples. Retaining all.")
        sampled_group = group

    sampled_data_list.append(sampled_group)

maps_downsampled = pd.concat(sampled_data_list)

maps_downsampled = maps_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)


print("\nDownsampling Complete.")
print(f"Original Dataset Size: {len(maps_data)}")
print(f"Downsampled Dataset Size: {len(maps_downsampled)}")
print("\nClass Distribution in Downsampled Dataset:")
print(maps_downsampled['cell_label'].value_counts())
```

```python
# convert to df
df = pd.DataFrame(maps_downsampled)
y = df['cell_label']

X = df.drop('cell_label', axis = 1)
```

### Next we use a nested cross validation strategy with extensive hyperparamter ranges

```python
pipeline = Pipeline([
    ('classifier', RandomForestClassifier(random_state = 10, n_jobs=1))
])  
 
param_grid = [
    # without smote
    {
        'classifier__n_estimators': [50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325, 350, 375, 400],
        'classifier__max_depth': [5, 10, 15, 20, None],
        'classifier__min_samples_split': [2, 5, 7, 10],
        'classifier__min_samples_leaf': [1, 2, 3, 4, 5],
        'classifier__max_features': ['sqrt', 'log2']
    }
]
```

```python
# run CV
inner_cv = StratifiedKFold(n_splits = 3, shuffle = True, random_state=42)
outer_cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state=42)

outer_loop_scores = []

final_cv_results = None 
print("Starting Nested Cross-Validation...")


for i, (train_idx, test_idx) in enumerate(outer_cv.split(X, y)):
    print(f"\n--- Outer Fold {i+1}/5 ---")
    
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    grid_search = GridSearchCV(
        estimator = pipeline,
        param_grid = param_grid,
        scoring = 'f1_macro',
        cv = inner_cv, 
        verbose = 0,
        n_jobs = 1
    )
    
    start_time = time.time()
    result = grid_search.fit(X_train, y_train)
    end_time = time.time()
    inner_loop_time = end_time - start_time
    print(f'Inner Loop time elapsed: {inner_loop_time:.2f} seconds')
    
    # best reuslts of current inner loop
    final_cv_results = result.cv_results_ 

    best_model = result.best_estimator_
    y_pred = best_model.predict(X_test)
    score = f1_score(y_test, y_pred, average='macro')
    outer_loop_scores.append(score)
    
    print(f"Best Score in Inner Loop (CV F1-Macro): {result.best_score_:.4f}")
    print(f"Best Hyperparameters: {result.best_params_}")
    print(f"Score on Outer Test Set (F1-Macro): {score:.4f}")
```

```python
print("\n--- Nested Cross-Validation Complete ---")
print(f"Mean F1-Macro Score: {np.mean(outer_loop_scores):.4f}")
print(f"Standard Deviation: {np.std(outer_loop_scores):.4f}")
```

```python
print(f"Best Hyperparameters: {result.best_params_}")
```

```python
# all hyperparameter combinations, model preffered larger forests, but was minimally improved in some cases and risk of overfitting 
results_df = pd.DataFrame(final_cv_results)
results_df
```

```python
# comparison of tree depth and F1
param_name = [col for col in results_df.columns if col.startswith('param_classifier__max') and col != 'param_pipe__name'][0] # Find the first param column

plt.figure(figsize=(8, 6))
plt.scatter(results_df[param_name], results_df['mean_test_score'], marker='o') 
plt.xlabel(f'Hyperparameter: {param_name.replace("param_", "")}')
plt.ylabel('Mean F1-Macro Score (Inner CV)')
plt.title(f'Performance vs. {param_name.replace("param_", "")} (Last Outer Fold)')
plt.show()
```

### The small bosst in performance gain meant the models are fairly robust once they hit their 'best' performance

```python
param_1_name = 'param_classifier__max_depth' 
param_2_name = 'param_classifier__n_estimators'

scores = results_df.pivot_table(
    index=param_1_name, 
    columns=param_2_name, 
    values='mean_test_score'
)

plt.figure(figsize=(12, 12))
sns.heatmap(
    scores, 
    annot=True, 
    fmt=".2f", 
    cmap="viridis", 
    cbar_kws={'label': 'Mean F1-Macro Score'}
)
plt.title(f'Hyperparameter heatmap (Last Outer Fold)')
plt.xlabel(param_2_name.replace("param_classifier__", ""))
plt.ylabel(param_1_name.replace("param_classifier__", ""))
plt.show()
```

### Below used a downsample strategy to see how SMOTE could be applied to potentially improve classification in cases where a minority cell type is present

```python
MIN_SAMPLES = 40
TARGET_CLASS_LABEL = 0
TARGET_CLASS_SIZE = 15

sampled_data_list = []

grouped = maps_data.groupby('cell_label')

for label, group in grouped:
    class_size = len(group)

    if label == TARGET_CLASS_LABEL:
        if class_size > TARGET_CLASS_SIZE:
            sampled_group = group.sample(n=TARGET_CLASS_SIZE, random_state=42)
            print(f"Target Minority Class {label} downsampled to {TARGET_CLASS_SIZE} samples.")
        else:
            print(f"Target Minority Class {label} retained with {class_size} samples (already small).")
            sampled_group = group

    elif class_size >= MIN_SAMPLES:
        sampled_group = group.sample(n=MIN_SAMPLES, random_state=42)
    else:
        print(f"Warning: Class {label} only has {class_size} samples. Retaining all.")
        sampled_group = group

    sampled_data_list.append(sampled_group)

maps_downsampled = pd.concat(sampled_data_list)

# Optional: Shuffle the final dataset
maps_downsampled = maps_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)

# --- Verification ---
print("\nDownsampling Complete.")
print(f"Original Dataset Size: {len(maps_data)}")
print(f"Downsampled Dataset Size: {len(maps_downsampled)}")
print("\nClass Distribution in Downsampled Dataset:")
print(maps_downsampled['cell_label'].value_counts())
```

```python
# new cross validation with and without smote
pipeline = Pipeline([
    ('smote', SMOTE(random_state=10)), 
    ('classifier', RandomForestClassifier(random_state=10, n_jobs=1))
])
# narrowed for faster checking
N_ESTIMATORS_GRID = [125, 150, 175, 200, 225, 250]

param_grid = [
 
    {

        'smote': [SMOTE(random_state=10)], 
        'smote__k_neighbors': [2, 3, 4], 
        'classifier__n_estimators': N_ESTIMATORS_GRID,
        'classifier__max_depth': [5, 10, 15, None],
        'classifier__min_samples_split': [2, 5, 7, 10],
        'classifier__min_samples_leaf': [1, 2, 3],
        'classifier__max_features': ['sqrt', 'log2']
    },
    

        'smote': ['passthrough'], 
        'classifier__n_estimators': N_ESTIMATORS_GRID,
        'classifier__max_depth': [5, 10, 15, None],
        'classifier__min_samples_split': [2, 5, 7, 10],
        'classifier__min_samples_leaf': [1, 2, 3],
        'classifier__max_features': ['sqrt', 'log2']
    }
]
```

```python
# run CV
inner_cv = StratifiedKFold(n_splits = 3, shuffle = True, random_state=42)
outer_cv = StratifiedKFold(n_splits = 5, shuffle = True, random_state=42)

outer_loop_scores = []
final_cv_results = None 
print("Starting Nested Cross-Validation...")


for i, (train_idx, test_idx) in enumerate(outer_cv.split(X, y)):
    print(f"\n--- Outer Fold {i+1}/5 ---")
    
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    grid_search = GridSearchCV(
        estimator = pipeline,
        param_grid = param_grid,
        scoring = 'f1_macro',
        cv = inner_cv, 
        verbose = 0, # Set to 0 to keep output clean, but keep 1 if desired
        n_jobs = 1
    )
    
    start_time = time.time()
    result = grid_search.fit(X_train, y_train)
    end_time = time.time()
    inner_loop_time = end_time - start_time
    print(f'Inner Loop time elapsed: {inner_loop_time:.2f} seconds')
    
    # Store the results of the *current* inner loop
    final_cv_results = result.cv_results_ 

    best_model = result.best_estimator_
    y_pred = best_model.predict(X_test)
    score = f1_score(y_test, y_pred, average='macro')
    outer_loop_scores.append(score)
    
    print(f"Best Score in Inner Loop (CV F1-Macro): {result.best_score_:.4f}")
    print(f"Best Hyperparameters: {result.best_params_}")
    print(f"Score on Outer Test Set (F1-Macro): {score:.4f}")
```
