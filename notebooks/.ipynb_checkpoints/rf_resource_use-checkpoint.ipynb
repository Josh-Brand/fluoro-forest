### Timing the use of random forest vs MAPS as increasing dataset sizes

```python
# requires cell_annotation environment
import sys
import os
import io
import json
import importlib
import numpy as np
import collections
import scipy
import sklearn
import time

import seaborn as sns
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt


from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.ensemble import BalancedRandomForestClassifier
```

```python
os.chdir('..')

# this file will be resampled while validation set is maintained
maps_train = pd.read_csv('/../Maps/data/cell_phenotyping/train.csv') # install from maps package

os.getcwd()
```

```python
maps_train.shape # 114k training sizes, each loop will be random but repeadted 5x, just keep separate
```

```python
sample_sizes = [250, 500, 1000, 2000, 5000, 10000, 25000, 50000]
```

```python
results = {}
with open(os.path.join('../../other_data/run_times/resource_indices.json'), 'r') as f:
    results = json.load(f)
```

```python
for k,v in results.items():
    print(k)
```

```python
timing_records = []
for k,v in results.items():
    sample_size = int(k)
    for k2, v2, in v.items():
        print(f'num_samples: {sample_size} - {k2}')
        replicate_key = k2

        current_sample_indices = v2['train']
               
        X_train_fold = maps_train.iloc[current_sample_indices, :-1]
        y_train_fold = maps_train.iloc[current_sample_indices, -1]


        start_time = time.time()
        ##
        rf = RandomForestClassifier(n_estimators = 250, max_features = 'sqrt',
                                    max_depth = 10, min_samples_leaf = 2, 
                                    min_samples_split = 5, random_state = 5)
        end_time = time.time()
        ##
        io_time = end_time - start_time

        del start_time, end_time

        start_time = time.time()
        ##
        rf.fit(X_train_fold, y_train_fold)
        end_time = time.time()
        ##
        fit_time = end_time - start_time

        timing_records.append({
            'sample_size': sample_size,
            'replicate': replicate_key,
            'io_time': io_time, # Time to instantiate Trainer object
            'fit_time': fit_time      # Time to run Trainer.fit()
        })

timing_df = pd.DataFrame(timing_records)
```

```python
rf_times = pd.read_csv('../../other_data/run_times/rf_training_time.csv')
rf_times['model'] = 'fluoroforest'

maps_times = pd.read_csv('../../other_data/run_times/maps_training_time.csv')
maps_times['model'] = 'maps'
```

```python
combined = pd.concat([rf_times, maps_times], axis = 0)
combined.head
```

```python
fig, ax = plt.subplots(figsize=(4, 3))

sns.lineplot(data=combined, x='sample_size', y='fit_time', hue='model', ax=ax, palette = 'tab10', hue_order=['fluoroforest', 'maps'])
sns.scatterplot(data=combined, x='sample_size', y='fit_time', hue='model', legend=None, ax=ax)

ax.set_yscale("log")
ax.set_yticks([1, 10, 100])
plt.grid(True, which="major", ls="--", linewidth=0.5)

ax.set_xlabel('Sample size')
ax.set_ylabel('Fit time (minutes)')

plt.title('Model by sample size (five replicates)')
plt.tight_layout()
plt.show()
```
